```python
import os
import numpy as np
import random

def load_random_mel_spectrogram(directory):
    """
    Загружает случайную мел-спектрограмму из указанной директории и выводит её размер.

    :param directory: Путь к директории с мел-спектрограммами.
    :return: Массив NumPy с мел-спектрограммой или None в случае ошибки.
    """
    try:
        # Получение списка файлов с расширением .npy
        files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f)) and f.endswith('.npy')]
        
        if not files:
            print("В указанной директории нет файлов с расширением .npy.")
            return None
        
        # Выбор случайного файла из списка
        random_file = random.choice(files)
        file_path = os.path.join(directory, random_file)
        
        # Загрузка мел-спектрограммы
        mel_spectrogram = np.load(file_path)
        
        # Вывод размера мел-спектрограммы
        print(f"Загружена мел-спектрограмма из файла: {random_file}")
        print(f"Размер мел-спектрограммы: {mel_spectrogram.shape}")
        
        return mel_spectrogram
    
    except Exception as e:
        print(f"Произошла ошибка при загрузке мел-спектрограммы: {e}")
        return None

# Пример использования функции
directory = '/workspace/TTSproject/fastspeech/FastSpeech/data/mels'
mel = load_random_mel_spectrogram(directory)
```

```python
import os
import random
from pathlib import Path
import librosa
from IPython.display import Audio
```

```python
def load_and_play_random_wav(directory):
    """
    Загружает случайный WAV-файл из указанной директории, воспроизводит его и выводит частоту дискретизации.
    
    :param directory: Путь к директории с WAV-файлами.
    :return: Объект Audio для воспроизведения в Jupyter Notebook или None в случае ошибки.
    """
    try:
        # Преобразование пути к объекту Path
        dir_path = Path(directory)
        
        # Проверка существования директории
        if not dir_path.exists():
            print(f"Директория не существует: {directory}")
            return None
        
        # Получение списка всех WAV-файлов в директории
        wav_files = list(dir_path.glob('*.wav'))
        
        # Проверка наличия WAV-файлов
        if not wav_files:
            print("В указанной директории нет WAV-файлов.")
            return None
        
        # Выбор случайного WAV-файла
        selected_wav = random.choice(wav_files)
        print(f"Выбранный файл: {selected_wav.name}")
        
        # Загрузка аудио с сохранением оригинальной частоты дискретизации
        y, sr = librosa.load(selected_wav, sr=None)
        
        # Вывод частоты дискретизации
        print(f"Частота дискретизации: {sr} Гц")
        
        # Воспроизведение аудио
        return Audio(y, rate=sr)
    
    except Exception as e:
        print(f"Произошла ошибка: {e}")
        return None
```

```python
# Укажите путь к директории с WAV-файлами
directory = '/workspace/TTSproject/fastspeech/FastSpeech/data/LJSpeech-1.1/wavs/'

# Вызов функции и воспроизведение аудио
audio_player = load_and_play_random_wav(directory)

# Отображение аудиоплеера в Jupyter Notebook
if audio_player:
    display(audio_player)
```

### функция для конвертации датасета в 16кгц для наших тестов

```python
import os
import librosa
import soundfile as sf

# Путь к директории с WAV-файлами
directory = '/workspace/TTSproject/fastspeech/FastSpeech/data/LJSpeech-1.1/wavs/'
i = 0
# Проход по всем файлам в директории
for root, _, files in os.walk(directory):
    for file in files:
        if file.lower().endswith('.wav'):
            filepath = os.path.join(root, file)
            try:
                # Загрузка аудио с оригинальной частотой дискретизации
                y, sr = librosa.load(filepath, sr=22050)
                
                # Конвертация частоты дискретизации на 16000 Гц
                y_resampled = librosa.resample(y, orig_sr=sr, target_sr=16000)
                
                # Сохранение аудио с новой частотой дискретизации, перезапись оригинала
                sf.write(filepath, y_resampled, 16000)
                
                if i%100 == 0:
                    print(f"Конвертирован: {filepath}")
            except Exception as e:
                print(f"Ошибка при конвертации {filepath}: {e}")
        i+= 1
```

### v2 через sox  - должно быть быстрее, протестировать в след раз

```python
import os
import subprocess
from pathlib import Path

def convert_wavs_to_16k_sox(directory):
    """
    Конвертирует все WAV-файлы в указанной директории с частотой дискретизации 22050 Гц на 16000 Гц,
    перезаписывая оригинальные файлы.

    :param directory: Путь к директории с WAV-файлами.
    """
    directory = Path(directory)
    
    if not directory.exists():
        print(f"Директория не существует: {directory}")
        return
    
    # Поиск всех WAV-файлов в директории и поддиректориях
    wav_files = list(directory.rglob('*.wav'))
    
    if not wav_files:
        print("В указанной директории нет WAV-файлов для конвертации.")
        return
    
    print(f"Найдено {len(wav_files)} WAV-файлов для конвертации.")
    
    for wav_file in wav_files:
        original_path = wav_file
        temp_path = wav_file.with_suffix('.temp.wav')
        
        try:
            # Выполнение команды SoX для конвертации частоты дискретизации
            subprocess.run([
                'sox', str(original_path),
                '-r', '16000',
                str(temp_path)
            ], check=True, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
            
            # Замена оригинального файла новым конвертированным файлом
            os.replace(temp_path, original_path)
            
            print(f"Конвертирован: {original_path}")
        
        except subprocess.CalledProcessError:
            print(f"Ошибка при конвертации файла: {original_path}")
            if temp_path.exists():
                temp_path.unlink()  # Удаление временного файла в случае ошибки
        except Exception as e:
            print(f"Неизвестная ошибка при обработке {original_path}: {e}")
            if temp_path.exists():
                temp_path.unlink()  # Удаление временного файла в случае ошибки

# Укажите путь к директории с WAV-файлами
directory = '/workspace/TTSproject/fastspeech/FastSpeech/data/LJSpeech-1.1/wavs/'

# Вызов функции для конвертации файлов
convert_wavs_to_16k_sox(directory)
```
